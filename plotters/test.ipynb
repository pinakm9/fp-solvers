```python
# add folders to Python's search space
import os, sys
from pathlib import Path
script_dir = Path(os.path.dirname(os.path.abspath('')))
module_dir = str(script_dir)
sys.path.insert(0, module_dir + '/modules')
print(module_dir)

# import the rest of the modules
%matplotlib nbagg
%matplotlib inline
import numpy as np
import tensorflow as tf 
import matplotlib.pyplot as plt
import arch
import pandas as pd
import tensorflow_probability as tfp
import time  
import sim
from mpl_toolkits.axes_grid1 import make_axes_locatable
import integrator as it 
DTYPE = 'float32'
```

```python
dim = 3
net = arch.LSTMForgetNet(50, 3, tf.float32, name="L63")
net.load_weights('../non-grad3D/data/L63/1M/{}'.format(net.name)).expect_partial()
X = tf.random.uniform(shape=(10, dim))
net(*tf.split(X, dim, axis=-1))
```

```python
o = np.ones(2, dtype=DTYPE)
mix = 0.5
comps = [tfp.distributions.MultivariateNormalDiag(loc=o, scale_diag=o),\
         tfp.distributions.MultivariateNormalDiag(loc=-o, scale_diag=o)]
cat = tfp.distributions.Categorical(probs=[mix, 1.-mix])
X = tfp.distributions.Mixture(cat=cat, components=comps)
```

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
Z = np.concatenate([x.reshape(-1, 1), y.reshape(-1, 1)], axis=-1)
ax.plot_surface(x, y, X.prob(Z).numpy().reshape(x.shape))
plt.show()
```
