```python
# add folders to Python's search space
import os, sys
from pathlib import Path
script_dir = Path(os.path.dirname(os.path.abspath('')))
module_dir = str(script_dir)
sys.path.insert(0, module_dir + '/modules')
print(module_dir)

# import the rest of the modules
%matplotlib nbagg
import numpy as np
import tensorflow as tf 
import matplotlib.pyplot as plt
import arch
import pandas as pd
import tensorflow_probability as tfp
import time  
import sim
import compare
from scipy.special import erf
DTYPE = 'float32'
```

```python
# load the learned solution
dim = 10
net = arch.LSTMForgetNet(50, 3, tf.float32, name="circle{}D".format(dim))
net.load_weights('../circle-fp/data/{}D/460/{}'.format(dim, net.name)).expect_partial()
X = tf.random.uniform(shape=(10, dim))
net(*tf.split(X, dim, axis=-1))

# define the true solution
D = 1.0
def p_inf(x, y):
  Z = 0.5 * np.sqrt(np.pi**3 * D) * (1. + erf(1/np.sqrt(D)))
  return tf.exp(-(x**2 + y**2 - 1.)**2 / D) / Z
```

```python
# set up plotting parameters
scale = 0
xlabel_size = ylabel_size = 15 + scale
tick_size = 10 + scale
legend_size = 15 + scale
title_size = 15 + scale
cbar_tick_size = 10 + scale
line_color = "darkgrey"

# plot solutions
def plot_solutions(learned, true, filename, resolution=50, low=[-2., -2], high=[2., 2.]):
  fig = plt.figure(figsize=(10, 5))
  ax_l = fig.add_subplot(121, projection='3d')
  ax_t = fig.add_subplot(122, projection='3d')
  # plotting params  
  ax_l.tick_params(axis='both', which='major', labelsize=tick_size)
  ax_l.tick_params(axis='both', which='minor', labelsize=tick_size)
  ax_l.set_title(r'Learned $p(0, 0, 0, 0, x_4, x_5, 0, 0, 0, 0)$', fontsize=title_size)
  ax_l.set_xlabel(r'$x_4$', fontsize=xlabel_size)
  ax_l.set_ylabel(r'$x_5$', fontsize=ylabel_size)
    
  ax_t.tick_params(axis='both', which='major', labelsize=tick_size)
  ax_t.tick_params(axis='both', which='minor', labelsize=tick_size)
  ax_t.set_title(r'True $p(0, 0, 0, 0, x_4, x_5, 0, 0, 0, 0)$', fontsize=title_size)
  ax_t.set_xlabel(r'$x_4$', fontsize=xlabel_size)
  ax_t.set_ylabel(r'$x_5$', fontsize=ylabel_size)
    
  delta = (high[0] - low[0]) * (high[1] - low[1]) / (resolution-1)**2
  
  x = np.linspace(low[0], high[0], num=resolution, endpoint=True).astype(DTYPE)
  y = np.linspace(low[1], high[1], num=resolution, endpoint=True).astype(DTYPE)
  y = np.repeat(y, resolution, axis=0).reshape((-1, 1))
  x = np.array(list(x) * resolution).reshape((-1, 1))
  z_l = learned(x, y).numpy()
  z_l /= (z_l.sum() * delta) 
  z_t = true(x, y).numpy()
  grid = (resolution, resolution)
  x = x.reshape(grid)
  y = y.reshape(grid)
  z_l = z_l.reshape(grid)
  z_t = z_t.reshape(grid)
  ax_l.plot_wireframe(x, y, z_l, color='red', linewidth=0.7)
  ax_t.plot_wireframe(x, y, z_t, color='blue', linewidth=0.7)
  # Get the bounding boxes of the axes including text decorations
  x = 0.475
  line = plt.Line2D([x+.045,x+.045],[0.0,1.0], transform=fig.transFigure, color=line_color)
  fig.add_artist(line)
  fig.tight_layout()
  plt.savefig('{}.png'.format(filename), dpi=300)
  plt.show()
  
def learned(x, y):
    z = tf.zeros_like(x)
    return tf.exp(net(z, z, z, z, x, y, z, z, z, z))

plot_solutions(learned=learned, true=p_inf, filename='../plots/10D-surface')
```
