```python
# add modules folder to Python's search path
import os, sys
from pathlib import Path
script_dir = Path(os.path.dirname(os.path.abspath('')))
module_dir = str(script_dir)
sys.path.insert(0, repo + '/modules')
print(module_dir)

# import the rest of the modules
%matplotlib nbagg
%matplotlib inline
import numpy as np
import tensorflow as tf 
import matplotlib.pyplot as plt
import arch
import tensorflow_probability as tfp
import time  
import sim
import integrator
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.animation as animation
from mpl_toolkits.axes_grid1 import make_axes_locatable
from scipy.special import erf
tfd = tfp.distributions
```

**Set computation and plotting parameters**

```python
# set up computation parameters
dim = 2
n_particles = int(8e7)
n_subdivs = 100
save_folder = '{}/ring-fp/data/2D'.format(repo)
n_steps = 1000
dt = 0.01
t = dt * n_steps
DTYPE = 'float32'

# set up plotting parameters
scale = 15
xlabel_size = ylabel_size = 15 + scale
tick_size = 10 + scale
legend_size = 15 + scale
title_size = 15 + scale
cbar_tick_size = 10 + scale
```

**Define the problem: $\mu, \sigma, p_0$**

```python
def mu(X):
    x, y = np.split(X, dim, axis=-1)
    z = -4. * (x*x + y*y - 1.0)
    return np.concatenate([x*z, y*z], axis=-1)

sigma = np.sqrt(2., dtype=DTYPE)
D = sigma**2 / 2.

l = np.ones(dim, dtype=DTYPE)
g1 = tfd.MultivariateNormalDiag(loc=l, scale_diag=l)
g2 = tfd.MultivariateNormalDiag(loc=-l, scale_diag=l)
mix = 0.5
rv0 = tfd.Mixture(cat=tfd.Categorical(probs=[mix, 1.-mix]), components=[g1, g2])
log_p0 = lambda x: tf.reshape(rv0.log_prob(x), (-1, 1))
```

**Set up Monte-Carlo solver**

```python
X0 = rv0.sample(n_particles).numpy()
mc_prob = sim.MCProb(save_folder, n_subdivs, mu, sigma, X0)
mc_prob.ready(n_steps=n_steps, dt=dt, lims=None)
```

**Load the neural network**

```python
n_theta = arch.LSTMForgetNet(50, 3, tf.float32, name="ring2D")
n_theta.load_weights('{}/{}'.format(save_folder, n_theta.name)).expect_partial()
X = tf.random.uniform(shape=(10, dim))
n_theta(*tf.split(X, dim, axis=-1))
```

**Define the true solution**

```python
Z = 0.5 * np.sqrt(np.pi**3 * D) * (1. + erf(1/np.sqrt(D, dtype=DTYPE)))
def p_inf(x, y):
  return np.exp(-(x**2 + y**2 - 1.)**2 / D) / Z
```

**Compute and save errors**

```python
grid = mc_prob.get_grid()
low = grid.mins
high = grid.maxs

x = np.linspace(low[0], high[0], num=n_subdivs+1)[:-1].astype(DTYPE) + grid.h[0]/2. 
y = np.linspace(low[1], high[1], num=n_subdivs+1)[:-1].astype(DTYPE) + grid.h[1]/2.
np.save('{}/x.npy'.format(save_folder), x)
np.save('{}/y.npy'.format(save_folder), y)

x_, y_ = list(map(lambda a: a.reshape(-1, 1), np.meshgrid(x, y)))


p_m = mc_prob.compute_p2(0, 1, save=False)
p_m *= (grid.h[0] * grid.h[1])
p_true = p_inf(x_, y_).reshape(p_m.shape)
p_true /= (p_true.sum())
p_l = np.exp(n_theta(x_, y_)).reshape(p_m.shape)
p_l /= (p_l.sum()) 

error_m = (p_m - p_true)
error_l = (p_l - p_true)

np.save('{}/p_m.npy'.format(save_folder), p_m)
np.save('{}/p_true.npy'.format(save_folder), p_true)
np.save('{}/p_l.npy'.format(save_folder), p_l)
np.save('{}/error_m.npy'.format(save_folder), error_m)
np.save('{}/error_l.npy'.format(save_folder), error_l)
```

**Plot errors**

```python
# set up plotting parameters
scale = 2
xlabel_size = ylabel_size = 15 + scale
tick_size = 10 + scale
legend_size = 15 + scale
title_size = 15 + scale
cbar_tick_size = 10 + scale
line_color = "darkgrey"

# define 6 subplots
fig_all = plt.figure(figsize=(20, 10))
ax_1l = fig_all.add_subplot(121) 
ax_1m = fig_all.add_subplot(122)


# position colorbars
div_1 = make_axes_locatable(ax_1l)
cax_1 = div_1.append_axes('right', '5%', '5%')
div_4 = make_axes_locatable(ax_1m)
cax_4 = div_4.append_axes('right', '5%', '5%')


# set label size
ax_1l.tick_params(axis='both', which='major', labelsize=tick_size)
ax_1l.tick_params(axis='both', which='minor', labelsize=tick_size)
ax_1m.tick_params(axis='both', which='major', labelsize=tick_size)
ax_1m.tick_params(axis='both', which='minor', labelsize=tick_size)


# write titles and labels
ax_1l.set_title('Deep learning error', fontsize=title_size)
ax_1m.set_xlabel(r'$x$', fontsize=xlabel_size)
ax_1m.set_ylabel(r'$y$', fontsize=ylabel_size)
ax_1m.set_title('Monte Carlo error at time {:.2f}'.format(t), fontsize=title_size)
ax_1m.set_xlabel(r'$x$', fontsize=xlabel_size)
ax_1m.set_ylabel(r'$y$', fontsize=ylabel_size)
    
im = ax_1m.pcolormesh(x, y, np.abs(error_m).T, cmap='inferno', shading='auto')
cbar_1m = fig_all.colorbar(im, cax=cax_4, ax=ax_1m)
cbar_1m.ax.tick_params(labelsize=cbar_tick_size)
ax_1m.set_aspect("auto")


im = ax_1l.pcolormesh(x, y, np.abs(error_l).T, cmap='inferno', shading='auto')
cbar_1l = fig_all.colorbar(im, cax=cax_1, ax=ax_1l)
cbar_1l.ax.tick_params(labelsize=cbar_tick_size)
ax_1l.set_aspect("auto")


# plot and save
fig_all.subplots_adjust(wspace=0.4, hspace=0.2)
# Get the bounding boxes of the axes including text decorations
s = 0.475
line = plt.Line2D([s+.045,s+.045],[0.08,0.9], transform=fig_all.transFigure, color=line_color)
fig_all.add_artist(line)
plt.savefig('{}/plots/2D-error.png'.format(repo))
```
